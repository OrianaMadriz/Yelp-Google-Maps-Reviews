```python
#convinamos los archivos
dfc=pd.concat([df,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15,df16,df17,df18], axis=0,join='inner')
```

```python
#leemos el archivo de la metadata
ms= pd.read_json('metadata-sitios/1.json',lines=True)
```

```python
#convinamos la metadata con el archivo original
merged_df = pd.merge(dfc, ms, on='gmap_id')
```

```python
#borramos las columnas que no sirven para la operaci√≥n.
df = merged_df.drop(['pics', 'resp', 'url'], axis=1)
df.head()
```

```python
#contamos los nulos
null_counts = merged_df.isnull().sum()
print(null_counts)
```

```python
#breve descripci√≥n de las caracteristicas de las columnas.
df.describe()
```

```python
#volvemos la columna categoria en una lista para poder operarla, y filtrar el dataframe con la categoria de restaurante.
df = df.explode('category')
df = df.dropna(subset=['category'])
df_restaurants = df[df['category'].str.contains('restaurant', case=False)]
```

```python
#top de los restauurates con mejores calificaciones
media_rating = df_restaurants.groupby('name_y')['rating'].mean()
top_15 = media_rating.sort_values(ascending=False).head(15)
print(top_15)
```

```python
conteo_gmap_id = df_restaurants['name_y'].value_counts()
top_10_gmap_id = conteo_gmap_id.nlargest(10)
```

```python
#Grafica de los restaurantes mas visitados.
top_10_gmap_id.plot.bar()
plt.xlabel('gmap_id')
plt.ylabel('N√∫mero de visitas')
plt.show()
```

```python
#las fechas en las hubo mayor numero de visitas a los restauranes.
df_restaurants['time'] = pd.to_datetime(df_restaurants['time'])
conteo_gmap_id = df_restaurants['name_y'].value_counts()
top_10_gmap_id = conteo_gmap_id.nlargest(10)
df_filtrado = df_restaurants[df_restaurants['name_y'].isin(top_10_gmap_id.index)]
conteo_visitas = df_filtrado.groupby(['name_y', 'time']).size()
fecha_max_visitas = conteo_visitas.groupby(level=0).idxmax()
fecha_max_visitas = fecha_max_visitas.reset_index(name='Fecha con el mayor n√∫mero de visitas')
fecha_max_visitas.columns = ['name_y', 'Fecha con el mayor n√∫mero de visitas']
fecha_max_visitas['Fecha con el mayor n√∫mero de visitas'] = fecha_max_visitas['Fecha con el mayor n√∫mero de visitas'].apply(lambda x: x[1] if isinstance(x, tuple) else x)

print(fecha_max_visitas)
```
